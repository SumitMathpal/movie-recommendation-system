```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
stopwords = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text): 
    text = str(text).lower()
    text = re.sub(r'[^a-zA-z\s]','',text) #remove the panctuation
    words = text.split()
    #remove sstopwords
    words = [word for word in words if word not in stopwords]
    #lematize 
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
    
    
```

```python
df["tags"] = df["tags"].apply(preprocess_text)
```

```python
df["tags"][1]
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index= df["title"]).drop_duplicates()
```

```python
indices
```

```python
#Vectorization TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
TFIDF = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words="english")
```

```python
TFIDF_matrix = TFDF.fit_transform(df["tags"])
```

```python
TFIDF_matrix.shape
```

```python
TFIDF_matrix
```

```python
#cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity

```

```python
 def recommed(title, n=10):
     if title not in indices: 
         return ["Movie Not Found"]
     idx = indices[title]
     sim_score = cosine_similarity(TFIDF_matrix[idx],TFIDF_matrix).flatten()
     similar_idx = sim_score.argsort()[::-1][1:n+1]
     return  df["title"].iloc[similar_idx]
```

```python
print(recommed("Avenger"))
```
